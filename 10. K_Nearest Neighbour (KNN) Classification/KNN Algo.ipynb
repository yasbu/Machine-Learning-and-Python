```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data.drop(['id','Unnamed: 32'],axis=1,inplace=True)
data.tail()
# malignant = M kötü huylu tumor
# benign = B iyi huylu tumor
```

```python
M = data[data.diagnosis=='M']
B = data[data.diagnosis=='B']
```

```python
M.info()
```

```python
# scatter plot
plt.scatter(M.radius_mean,M.area_mean,color='red',label='bad')
plt.scatter(B.radius_mean,B.area_mean,color='green',label='good')
plt.legend()
plt.show()
```

```python
# scatter plot
plt.scatter(M.radius_mean,M.texture_mean,color='red',label='bad')
plt.scatter(B.radius_mean,B.texture_mean,color='green',label='good')
plt.xlabel('radius_mean')
plt.ylabel('texture_mean')
plt.legend()
plt.show()
```

```python
data.diagnosis = [1 if i == 'M' else 0 for i in data.diagnosis]

y = data.diagnosis.values
x_data = data.drop(['diagnosis'],axis=1)
```

```python
x_data
```

```python
# normalization

x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
x
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) # n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

```python
prediction
```

```python
print('{} nn score: {}'.format(3, knn.score(x_test,y_test)))
```

```python
# find k value
score_list=[]
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors= each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))

plt.plot(range(1,15),score_list)
plt.xlabel('k values')
plt.ylabel('accuracy')
plt.show()
```

## Conclusion
* k value eğer 8 değerine sahipse benim test sonuçların en iyi değeri veriyor.
